# Calculate Net Present Value

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, timedelta
from dateutil.relativedelta import relativedelta

# https://www.codingfinance.com/post/2018-03-19-tvm_py/
```

```python
def discount(t, r):
    """
    Compute the price of a pure discount bond that pays $1 at time t 
    where t is in years and r is the annual interest rate
    """
    return (1+r)**(-t)

def presentvalue(l, r):
    """
    Compute the present value of a list of cash flow given by the time (as an index) 
    and amounts l in a dataframe indexed by time
    """
    dates = l.index
    discounts = discount(dates, r)
    return (discounts*l)

def net_pv(l,r):
    """
    Compute the net present value of a list of cash flow given by the time (as an index) 
    and amounts l in a dataframe indexed by time
    net present value is the sum of present value
    """
    return presentvalue(l,r).sum()

def myplot(xvalue, yvalue, title="no title", xlabel="years", ylabel="cashflow", printTable=True):
    if printTable:
        print(xvalue)
    
    # plot the  present value
    fig = plt.figure(figsize=(12,6))
    ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
    ax1.plot(xvalue, yvalue)
    ax1.set_xlabel(xlabel) # Notice the use of set_ to begin methods
    ax1.set_ylabel(ylabel)
    ax1.set_title(title)
    plt.show()

def myplot2(xyvalue, title="no title", xlabel="years", ylabel="cashflow", printTable=True):
    
    if printTable:
        print(xyvalue)
    
    # plot the  present value
    fig = plt.figure(figsize=(12,6))
    ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
    ax1.plot(xyvalue.index, xyvalue['present_value'])
    ax1.set_xlabel(xlabel) # Notice the use of set_ to begin methods
    ax1.set_ylabel(ylabel)
    ax1.set_title(title)
    plt.show()
```

# Calculate the net present value of equal cash flow 

```python
# dataframe of liabilites and when it is owed..example $100 owed 1yrs from now
# $100 owed 2 years from now....
discount_rate = .1
cashflow = pd.Series(data=[100,100,100,100,100], index=[1,2,3,4,5])
cashflow = cashflow.rename('cash_flow')

df = pd.DataFrame(cashflow)
df.insert(1, "present_value", presentvalue(cashflow,discount_rate))

myplot2(df, title="present value")
```

# Calculating net present value of uneven cash flows

```python
# dataframe of liabilites and when it is owed..example $100 owed 1yrs from now
# $100 owed 2 years from now....
discount_rate = .1
cashflow = pd.Series(data=[100,200,500,300,100], index=[1,2,3,4,5])
#cashflow = pd.Series(data=[100], index=[30])
print(presentvalue(cashflow,discount_rate))
print("the net present value is {}".format(net_pv(cashflow,discount_rate)))

# plot the  present value
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(cashflow.index, presentvalue(cashflow,discount_rate))
ax1.set_xlabel('Years') # Notice the use of set_ to begin methods
ax1.set_ylabel('PV')
ax1.set_title('Present Value Chart')
plt.show()
```

# Calculating the PV with negative cash flow

In many cases we will have an initial cash flow that is negative. Think of this as buying a bond, where we buy a bond for $250 and it pays an interest of $100 each year. We can solve the present value of such a bond as below

```python
# dataframe of liabilites and when it is owed..example $100 owed 1yrs from now
# $100 owed 2 years from now....
discount_rate = .1
cashflow = pd.Series(data=[-250,100,100,100,100,350], index=[0,1,2,3,4,5])
#cashflow = pd.Series(data=[100], index=[30])
print(presentvalue(cashflow,discount_rate))
print("the net present value is {}".format(net_pv(cashflow,discount_rate)))

# plot the  present value
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(cashflow.index, presentvalue(cashflow,discount_rate))
ax1.set_xlabel('Years') # Notice the use of set_ to begin methods
ax1.set_ylabel('PV')
ax1.set_title('Present Value Chart')
plt.show()
```

# Calculating the present value of a finite annuity

In some cases we have to calculate the present value of stream of equal payments. Many times a lottery winner is given the option to get the money in equal payments or a lump sum. Let us suppose you win a $50000 lottery and you are given two options

Receive five equal payments of $10000
Lump sum $45000.

Let us suppose the bank pays 5% interest rate on the deposit.

At first glance it may appear that five equal payments of $10000 is better than $45000. After all $50000 is more than $45000. To solve this problem we can calculate the net present value of those payments as below.

```python
# dataframe of liabilites and when it is owed..example $100 owed 1yrs from now
# $100 owed 2 years from now....
lumpsum = 50000
discount_rate = .05
cashflow = pd.Series(data=[10000,10000,10000,10000,10000], index=[1,2,3,4,5])
#cashflow = pd.Series(data=[100], index=[30])
npv = net_pv(cashflow,discount_rate)


print("The net present value of five equal payments is ${0:.2f}".format(npv))

if npv < lumpsum:
    print("This payment is less than ${} offered today".format(lumpsum))
    print("Therefore it is better to take the lump sum payment")
elif npv > lumpsum:
    print("This payment is more than ${} offered today".format(lumpsum))
    print("Therefore it is better to take the equal payments") 
else:
    print("Both are same. Take anything")
```
