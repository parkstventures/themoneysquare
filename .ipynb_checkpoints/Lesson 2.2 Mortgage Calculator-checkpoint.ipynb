```python
# Ref: https://medium.com/towards-artificial-intelligence/mortgage-calculator-python-code-94d976d25a27
# Dated: Mar 13, 2020
import pandas as pd
import numpy as np
from datetime import date
import matplotlib.pyplot as plt
import matplotlib
import numpy_financial as npf
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
# this is final version based on 
# ref - https://github.com/chris1610/pbpython/blob/master/notebooks/Amortization-Model-Article.ipynb

def Cal_Mortgage(Principal, Interest_Rate, Mortgage_Duration, Payments_Year, Additional_Principal=0, Start_Date=date.today()):
    """ Calculate the amortization schedule given the loan details
    
     Args:
        Principal: Amount borrowed
        Interest_Rate: The annual interest rate for this loan
        Mortgage_Duration: Number of years for the loan
        Payments_Year: Number of payments in a year
        
        Additional_Principal (optional): Additional payments to be made each period. Assume 0 if nothing provided.
                                   must be a value less then 0, the function will convert a positive value to
                                   negative
        Start_Date (optional): Start date. Will start on first of next month if none provided

    Returns:
        schedule: Amortization schedule as a pandas dataframe
        summary: Pandas dataframe that summarizes the payoff information
    """
    # Compute actual loan amount by subtracting the downpayment
    # Principal = SalesPrice * ( 1 - DownPayment)

    Normalized_Duration = Mortgage_Duration * Payments_Year
    Normalized_Interest_Rate = Interest_Rate / (Payments_Year * 100)

    # Ensure the additional payments are negative
    if Additional_Principal > 0:
        Additional_Principal = -Additional_Principal
    
    # Create an index of the payment dates
    rng = pd.date_range(Start_Date, periods=Normalized_Duration, freq='MS')
    rng.name = P4
    
    # Build up the Amortization schedule as a DataFrame
    df = pd.DataFrame(index=rng,columns=["Payment", "Principal", 'Interest', 
                                         "Extra Principal", 'CurrentBalance'], dtype='float')
    
    # Add index by period (start at 1 not 0)
    df.reset_index(inplace=True)
    df.index += 1
    df.index.name = "Period"
    
    # Calculate the payment, principal and interests amounts using built in Numpy functions
    per_payment = npf.pmt(Normalized_Interest_Rate, Normalized_Duration, Principal)
    df["Payment"] = per_payment
    df["Principal"] = npf.ppmt(Normalized_Interest_Rate, df.index, Normalized_Duration, Principal)
    df["Interest"] = npf.ipmt(Normalized_Interest_Rate, df.index, Normalized_Duration, Principal)
        
    # Round the values
    df = df.round(2) 
    
    # Add in the additional principal payments
    df["Extra Principal"] = Additional_Principal
    
    # Store the Cumulative Principal Payments and ensure it never gets larger than the original principal
    df["Cumulative Interest"] = (df["Interest"]).cumsum()

    df["Cumulative Principal"] = (df["Principal"] + df["Extra Principal"]).cumsum()
    df["Cumulative Principal"] = df["Cumulative Principal"].clip(lower=-Principal)
    
    # Calculate the current balance for each period
    df["CurrentBalance"] = Principal + df["Cumulative Principal"]
    
    # Determine the last payment date
    try:
        last_payment = df.query("CurrentBalance <= 0")["CurrentBalance"].idxmax(axis=1, skipna=True)
    except ValueError:
        last_payment = df.last_valid_index()
    
    last_payment_date = "{:%m-%d-%Y}".format(df.loc[last_payment, "Payment Date"])
        
    # Truncate the data frame if we have additional principal payments:
    if Additional_Principal != 0:
                
        # Remove the extra payment periods
        df = df.loc[0:last_payment].copy()
        
        # Calculate the principal for the last row
        df.loc[last_payment, "Principal"] = -(df.loc[last_payment-1, "CurrentBalance"])
        
        # Calculate the total payment for the last row
        df.loc[last_payment, "Payment"] = df.loc[last_payment, ["Principal", "Interest"]].sum()
        
        # Zero out the additional principal
        df.loc[last_payment, "Extra Principal"] = 0
        
    # Get the payment info into a DataFrame in column order
    payment_info = (df[["Payment", "Principal", "Extra Principal", "Interest"]]
                    .sum().to_frame().T)
       
    # Format the Date DataFrame
    payment_details = pd.DataFrame.from_dict(dict([('Payoff Date', [last_payment_date]),
                                               ('Interest Rate', [Interest_Rate]),
                                               ('Number of Years', [Mortgage_Duration])
                                              ]))
    # Add a column showing how much we pay each period.
    # Combine addl principal with principal for total payment
    payment_details["Period Payment"] = round(per_payment, 2) + Additional_Principal
    
    payment_summary = pd.concat([payment_details, payment_info], axis=1)
    return df, payment_summary

```

```python
SalesPrice = 120000
# downpayment in %
DownPayment = .1
# InterestRate in %
Interest_Rate = 4.75
#Number of years
Mortgage_Duration=30
# number of payments in year
Payments_Year=12

Principal = SalesPrice * ( 1 - DownPayment)
Additional_Principal=0
```

```python
# Calculate
schedule1, stats1 = Cal_Mortgage(Principal,Interest_Rate, Mortgage_Duration, Payments_Year,  Additional_Principal)
(schedule1, stats1)
```

```python
schedule1['Interest'].plot()
```

```python
# compare multiple scenarios
Additional_Principal=100

schedule2, stats2 = Cal_Mortgage(Principal,Interest_Rate, Mortgage_Duration, Payments_Year,  Additional_Principal)

Interest_Rate = 3.75
Additional_Principal=0

schedule3, stats3 = Cal_Mortgage(Principal,Interest_Rate, Mortgage_Duration, Payments_Year,  Additional_Principal)

pd.concat([stats1, stats2, stats3], ignore_index=True)
```
